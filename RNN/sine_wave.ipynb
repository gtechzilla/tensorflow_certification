```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating our sine wave
x = np.linspace(0,50,501)
# sine of each data point
y = np.sin(x)
#plot the wave
plt.plot(x,y)
```

```python
#convering to df
data_df = pd.DataFrame(data=y,index=x,columns=['Sine'])
data_df.head()
```
